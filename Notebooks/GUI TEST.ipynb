```python
class SampleApp(tk.Tk):

    def __init__(self, *args, **kwargs):
        # *args para argumentos y **kwargs para diccionarios de argumentos de palabras clave
        tk.Tk.__init__(self, *args, **kwargs)
        #__init__ para inicializar los atributos
        
        #creaccion del marco del programa
        container = tk.Frame(self)
        container.pack(side="top", fill="both", expand=True)
        #coloco las paginas de manera que se puedan recorrer y la primera sea la principal
        self.frames = {}
        for F in (Main, Predecir, Guia,Bitcoin,Ethereum,Cardano,MovingAverageA,MovingAverageB,MovingAverageE,SVM_ETH,SVM_BTC,SVM_ADA,RLM_BTC,RLM_ETH,RLM_ADA):
            #Recorre por marcos
            page_name = F.__name__
            #__name__ usado de constructor
            frame = F(parent=container, controller=self)
            self.frames[page_name] = frame
            frame.grid(row=0, column=0, sticky="nsew")
        #muestro el main por defecto    
        self.show_frame("Main")
        
        
    def show_frame(self, page_name):
        #marco para la pagina
        frame = self.frames[page_name]
        frame.tkraise()
        
    

class Main(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        #inicializacion para la clase Main
        #self = .!frame.!main
   

        #Funcion para salir de la aplicacion
        def ExitApplication():
            MsgBox = tk.messagebox.askquestion ('Salir de la aplicación',
                                                '¿Estás seguro que quieres salir de la aplicación?',icon = 'warning')
            if MsgBox == 'yes':
                app.destroy()
                #En el caso de confirmar cierra la aplicacion
            else:
                tk.messagebox.showinfo('Confirmación','Volverás a la pantalla inicial')
                #Opcion de volver en el caso de que haya sido un error
        
        #creacción del marco para la interfaz denotado con dimension 1000x500
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()

        #Etiqueta para mostar en el apartado principal
        Titulo = tk.Label(self, text='PROYECTO FIN DE GRADO')
        Titulo.config(font=('Times New Roman', 36))
        Nombre = tk.Label(self, text='Miguel Ángel Vara Martínez')
        Nombre.config(font=('Times New Roman', 20))
        canvas.create_window(500, 100, window=Titulo)
        canvas.create_window(500, 150, window=Nombre)
        
        #Botones pantalla del Main
        PredictButtom = tk.Button(self, text="Predecir",command=lambda: controller.show_frame("Predecir"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(250, 300, window=PredictButtom, height=100, width=180)
        
        
        GuideButton = tk.Button(self, text="Guia",command=lambda: controller.show_frame("Guia"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(500, 300, window=GuideButton, height=100, width=180)
        
        
        exit = tk.Button (self, text='Salir de la aplicación',command=ExitApplication,bg='brown',
                          fg='white',height = 5, width =17,font=('Arial', 11, 'bold'))
        canvas.create_window(750, 300, window=exit, height=100, width=180)
        
        

class Predecir(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        #creacción del marco para la interfaz denotado con dimension 1000x500
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        #Etiqueta para mostar en el apartado principal
        Titulo = tk.Label(self, text='Predecir')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        
        BitcoinButtom = tk.Button(self, text="Bitcoin",command=lambda: controller.show_frame("Bitcoin"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(250, 200, window=BitcoinButtom, height=100, width=180)
        
        
        CardanoButton = tk.Button(self, text="Cardano",command=lambda: controller.show_frame("Cardano"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=CardanoButton, height=100, width=180)
        
        EthereumButton = tk.Button(self, text="Ethereum",command=lambda: controller.show_frame("Ethereum"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(750, 200, window=EthereumButton, height=100, width=180)
        
               
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 350, window=button, height=100, width=180)
#######################################################################################################################   
        
#Clase para la Media Movil     
class MovingAverageA(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='Medias Móviles')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        Ecuacion = tk.Label(self,text='Medias Móviles aplicadas a 15 y 30 días')
        Ecuacion.config(font=('Arial', 18))
        canvas.create_window(500, 100, window=Ecuacion)
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Año')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Mes')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Día')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        
        def BBBB():
        #igualamos el valor introducido por pantalla
            x1=str(entry1.get())
            x2=str(entry2.get())
            x3=str(entry3.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            #la funcion plotloo esta en un script aparte
            outputt = FigureCanvasTkAgg(MMA(x1,x2,x3), master=self)
            #llamada a la funcion plotloob del script HISTCRYPTO
            outputt.get_tk_widget().pack()
            outputt.draw()
            if outputt:
                #funcion para limpiar la grafica
                def clearplotB():
                    limpiarMMbutton.destroy()
                    outputt.get_tk_widget().destroy()
                    #creo otro boton por si queremos crear, destruir continuamente
                    Predecir = tk.Button(self, text="Calcular",command =lambda: BBBB() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
                limpiarMMbutton = tk.Button(self, text="Limpiar",command = lambda:clearplotB(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(500, 300,window=limpiarMMbutton, height=50, width=180)
            else:
                print("error")
                    
        Predecir = tk.Button(self, text="Calcular",command =lambda: BBBB() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
 

    #Clase para la Media Movil     
class MovingAverageB(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='Medias Móviles')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        Ecuacion = tk.Label(self,text='Medias Móviles aplicadas a 15 y 30 días')
        Ecuacion.config(font=('Arial', 18))
        canvas.create_window(500, 100, window=Ecuacion)
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Año')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Mes')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Día')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        
        def BBBB():
        #igualamos el valor introducido por pantalla
            x1=str(entry1.get())
            x2=str(entry2.get())
            x3=str(entry3.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            #la funcion plotloo esta en un script aparte
            outputt = FigureCanvasTkAgg(MMB(x1,x2,x3), master=self)
            #llamada a la funcion plotloob del script HISTCRYPTO
            outputt.get_tk_widget().pack()
            outputt.draw()
            if outputt:
                #funcion para limpiar la grafica
                def clearplotB():
                    limpiarMMbutton.destroy()
                    outputt.get_tk_widget().destroy()
                    #creo otro boton por si queremos crear, destruir continuamente
                    Predecir = tk.Button(self, text="Calcular",command =lambda: BBBB() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
                limpiarMMbutton = tk.Button(self, text="Limpiar",command = lambda:clearplotB(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(500, 300,window=limpiarMMbutton, height=50, width=180)
            else:
                print("error")
                    
        Predecir = tk.Button(self, text="Calcular",command =lambda: BBBB() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        


#Clase para la Media Movil     
class MovingAverageE(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='Medias Móviles')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        Ecuacion = tk.Label(self,text='Medias Móviles aplicadas a 15 y 30 días')
        Ecuacion.config(font=('Arial', 18))
        canvas.create_window(500, 100, window=Ecuacion)
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Año')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Mes')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Día')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        
        def BBBB():
        #igualamos el valor introducido por pantalla
            x1=str(entry1.get())
            x2=str(entry2.get())
            x3=str(entry3.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            #la funcion plotloo esta en un script aparte
            outputt = FigureCanvasTkAgg(MME(x1,x2,x3), master=self)
            #llamada a la funcion plotloob del script HISTCRYPTO
            outputt.get_tk_widget().pack()
            outputt.draw()
            if outputt:
                #funcion para limpiar la grafica
                def clearplotB():
                    limpiarMMbutton.destroy()
                    outputt.get_tk_widget().destroy()
                    #creo otro boton por si queremos crear, destruir continuamente
                    Predecir = tk.Button(self, text="Calcular",command =lambda: BBBB() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
                limpiarMMbutton = tk.Button(self, text="Limpiar",command = lambda:clearplotB(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(500, 300,window=limpiarMMbutton, height=50, width=180)
            else:
                print("error")
                    
        Predecir = tk.Button(self, text="Calcular",command =lambda: BBBB() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
#######################################################################################################################    
        
        
#Clase para la Regresion Lineal Multiple     
class RLM_BTC(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='Regresion Lineal Multiple')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        Ecuacion = tk.Label(self,text='Y = b + m1 X1 + m2 X2 + … + m(n) X(n)')
        Ecuacion.config(font=('Arial', 18))
        canvas.create_window(500, 100, window=Ecuacion)
        
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Precio de Apertura')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)
        

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Dia')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Mes')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        entry4 = tk.Entry (self)
        canvas.create_window(250, 360, window=entry4) 
        anyolabel = tk.Label(self,text='Año')
        anyolabel.config(font=('Arial', 14))
        canvas.create_window(100, 360, window=anyolabel)
        
        def bbb():
        #igualamos el valor introducido por pantalla
            x1=float(entry1.get())
            x2=float(entry2.get())
            x3=float(entry3.get())
            x4=float(entry4.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            label1 = tk.Label(self, text= 'Cierre estimado en: ',font=('Arial', 16,'bold'))
            canvas.create_window(700, 300, window=label1)
            
            label2 = tk.Label(self, text= RLMBTC(x1,x2,x3,x4),font=('Arial', 16,'bold'))
            canvas.create_window(720, 350, window=label2)
            
            label3 = tk.Label(self, text= '€',font=('Arial', 16,'bold'))
            canvas.create_window(780, 350, window=label3)
            
            
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
        Predecir = tk.Button(self, text="Predecir",command =lambda: bbb() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        

  
        
        
#Clase para la Regresion Lineal Multiple     
class RLM_ADA(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='Regresion Lineal Multiple')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        Ecuacion = tk.Label(self,text='Y = b + m1 X1 + m2 X2 + … + m(n) X(n)')
        Ecuacion.config(font=('Arial', 18))
        canvas.create_window(500, 100, window=Ecuacion)
        
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Precio de Apertura')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)
        

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Dia')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Mes')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        entry4 = tk.Entry (self)
        canvas.create_window(250, 360, window=entry4) 
        anyolabel = tk.Label(self,text='Año')
        anyolabel.config(font=('Arial', 14))
        canvas.create_window(100, 360, window=anyolabel)
        
        def bbb():
        #igualamos el valor introducido por pantalla
            x1=float(entry1.get())
            x2=float(entry2.get())
            x3=float(entry3.get())
            x4=float(entry4.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            label1 = tk.Label(self, text= 'Cierre estimado en: ',font=('Arial', 16,'bold'))
            canvas.create_window(700, 300, window=label1)
            
            label2 = tk.Label(self, text= RLMADA(x1,x2,x3,x4),font=('Arial', 16,'bold'))
            canvas.create_window(720, 350, window=label2)
            
            label3 = tk.Label(self, text= '€',font=('Arial', 16,'bold'))
            canvas.create_window(780, 350, window=label3)
            
            
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
        Predecir = tk.Button(self, text="Predecir",command =lambda: bbb() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )

        
        
#Clase para la Regresion Lineal Multiple     
class RLM_ETH(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='Regresion Lineal Multiple')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        Ecuacion = tk.Label(self,text='Y = b + m1 X1 + m2 X2 + … + m(n) X(n)')
        Ecuacion.config(font=('Arial', 18))
        canvas.create_window(500, 100, window=Ecuacion)
        
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Precio de Apertura')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)
        

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Dia')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Mes')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        entry4 = tk.Entry (self)
        canvas.create_window(250, 360, window=entry4) 
        anyolabel = tk.Label(self,text='Año')
        anyolabel.config(font=('Arial', 14))
        canvas.create_window(100, 360, window=anyolabel)
        
        def bbb():
        #igualamos el valor introducido por pantalla
            x1=float(entry1.get())
            x2=float(entry2.get())
            x3=float(entry3.get())
            x4=float(entry4.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            label1 = tk.Label(self, text= 'Cierre estimado en: ',font=('Arial', 16,'bold'))
            canvas.create_window(700, 300, window=label1)
            
            label2 = tk.Label(self, text= RLMETH(x1,x2,x3,x4),font=('Arial', 16,'bold'))
            canvas.create_window(720, 350, window=label2)
            
            label3 = tk.Label(self, text= '€',font=('Arial', 16,'bold'))
            canvas.create_window(780, 350, window=label3)
            
            
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
        Predecir = tk.Button(self, text="Predecir",command =lambda: bbb() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        
####################################################################################################    



#Clase para la Regresion Lineal Multiple     
class SVM_ADA(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='SVM')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Precio de Apertura')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)
        

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Dia')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Mes')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        entry4 = tk.Entry (self)
        canvas.create_window(250, 360, window=entry4) 
        anyolabel = tk.Label(self,text='Año')
        anyolabel.config(font=('Arial', 14))
        canvas.create_window(100, 360, window=anyolabel)
        
        def bbbada():
        #igualamos el valor introducido por pantalla
            x1=float(entry1.get())
            x2=float(entry2.get())
            x3=float(entry3.get())
            x4=float(entry4.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            label1 = tk.Label(self, text= 'Cierre estimado en: ',font=('Arial', 16,'bold'))
            canvas.create_window(700, 300, window=label1)
            
            label2 = tk.Label(self, text= SVMADA(x1,x2,x3,x4),font=('Arial', 16,'bold'))
            canvas.create_window(720, 350, window=label2)
            
            label3 = tk.Label(self, text= '€',font=('Arial', 16,'bold'))
            canvas.create_window(780, 350, window=label3)
            
            
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
        Predecir = tk.Button(self, text="Predecir",command =lambda: bbbada() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        
        
        
        
#Clase para la Regresion Lineal Multiple     
class SVM_BTC(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='SVM')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Precio de Apertura')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)
        

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Dia')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Mes')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        entry4 = tk.Entry (self)
        canvas.create_window(250, 360, window=entry4) 
        anyolabel = tk.Label(self,text='Año')
        anyolabel.config(font=('Arial', 14))
        canvas.create_window(100, 360, window=anyolabel)
        
        def bbbbtc():
        #igualamos el valor introducido por pantalla
            x1=float(entry1.get())
            x2=float(entry2.get())
            x3=float(entry3.get())
            x4=float(entry4.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            label1 = tk.Label(self, text= 'Cierre estimado en: ',font=('Arial', 16,'bold'))
            canvas.create_window(700, 300, window=label1)
            
            label2 = tk.Label(self, text= SVMBTC(x1,x2,x3,x4),font=('Arial', 16,'bold'))
            canvas.create_window(720, 350, window=label2)
            
            label3 = tk.Label(self, text= '€',font=('Arial', 16,'bold'))
            canvas.create_window(780, 350, window=label3)
            
            
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
        Predecir = tk.Button(self, text="Predecir",command =lambda: bbbbtc() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        
        
        
#Clase para la Regresion Lineal Multiple     
class SVM_ETH(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='SVM')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        
        
        entry1 = tk.Entry (self)
        canvas.create_window(250, 240, window=entry1) 
        abrirlabel = tk.Label(self,text='Precio de Apertura')
        abrirlabel.config(font=('Arial', 14))
        canvas.create_window(100, 240, window=abrirlabel)
        

        entry2 = tk.Entry (self)
        canvas.create_window(250, 280, window=entry2) 
        dialabel = tk.Label(self,text='Dia')
        dialabel.config(font=('Arial', 14))
        canvas.create_window(100, 280, window=dialabel)

        entry3 = tk.Entry (self)
        canvas.create_window(250, 320, window=entry3) 
        meslabel = tk.Label(self,text='Mes')
        meslabel.config(font=('Arial', 14))
        canvas.create_window(100, 320, window=meslabel)
        
        entry4 = tk.Entry (self)
        canvas.create_window(250, 360, window=entry4) 
        anyolabel = tk.Label(self,text='Año')
        anyolabel.config(font=('Arial', 14))
        canvas.create_window(100, 360, window=anyolabel)
        
        def bbbeth():
        #igualamos el valor introducido por pantalla
            x1=float(entry1.get())
            x2=float(entry2.get())
            x3=float(entry3.get())
            x4=float(entry4.get())
            #LLamada a la funcion de regresion lineal multiple de cardano
            #LABEL para mostrar la prediccion
            label1 = tk.Label(self, text= 'Cierre estimado en: ',font=('Arial', 16,'bold'))
            canvas.create_window(700, 300, window=label1)
            
            label2 = tk.Label(self, text= SVMETH(x1,x2,x3,x4),font=('Arial', 16,'bold'))
            canvas.create_window(720, 350, window=label2)
            
            label3 = tk.Label(self, text= '€',font=('Arial', 16,'bold'))
            canvas.create_window(780, 350, window=label3)
            
            
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
        Predecir = tk.Button(self, text="Predecir",command =lambda: bbbeth() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Predecir,height=50, width=180 )
        
#################################################################################################### 

                    
#Clase para la Guia  
class Guia(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='Guia de Criptomonedas')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 50, window=Titulo)
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 350, window=button, height=100, width=180)

####################################################################################################    

#Clase Bitcoin  
        
class Bitcoin(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        #creacción del marco para la interfaz denotado con dimension 1000x500
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        #Etiqueta para mostar en el apartado principal
        Titulo = tk.Label(self, text='BITCOIN')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 40, window=Titulo)
        #funcion creada para graficar el historico asociado a esa cripto
        def pplot (): 
            #la funcion plotloo esta en un script aparte
            output = FigureCanvasTkAgg(plotloob(), master=self)
            #llamada a la funcion plotloob del script HISTCRYPTO
            output.get_tk_widget().pack()
            output.draw()
            #en el caso de tener el historico
            if output:
                #funcion para limpiar la grafica
                def clearplot():
                    limpiarbutton.destroy()
                    output.get_tk_widget().destroy()
                    #creo otro boton por si queremos crear, destruir continuamente
                    Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )
                    
                    #creo otro boton por si queremos crear, destruir continuamente
                    HistoricoMediabutton = tk.Button(self, text="Medias Móviles en el último año",command =lambda: plotMedias() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(750, 200, window=HistoricoMediabutton,height=50, width=250 )
                    

                limpiarbutton = tk.Button(self, text="Limpiar",command = lambda:clearplot(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(500, 300,window=limpiarbutton, height=50, width=180)
            else:
                print("error")
                
        Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )
        
        
        def plotMedias (): 
            #la funcion plotloo esta en un script aparte
            outputt = FigureCanvasTkAgg(pplotMediasBTC(), master=self)
            #llamada a la funcion plotloob del script HISTCRYPTO
            outputt.get_tk_widget().pack()
            outputt.draw()
            #en el caso de tener el historico
            if outputt:
                #funcion para limpiar la grafica
                def clearplotMedias():
                    limpiarMediasbutton.destroy()
                    outputt.get_tk_widget().destroy()
                    #creo otro boton por si queremos crear, destruir continuamente
                    HistoricoMediabutton = tk.Button(self, text="Medias Móviles en el último año",command =lambda: plotMedias() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(750, 200, window=HistoricoMediabutton,height=50, width=250 )
                    
                    #creo otro boton por si queremos crear, destruir continuamente
                    Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )

                limpiarMediasbutton = tk.Button(self, text="Limpiar",command = lambda:clearplotMedias(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(750, 300,window=limpiarMediasbutton, height=50, width=180)
            else:
                print("error")
                
        HistoricoMediabutton = tk.Button(self, text="Medias Móviles en el último año",command =lambda: plotMedias() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(750, 200, window=HistoricoMediabutton,height=50, width=250 )
        
        #Boton para llamar a la MediaMovil
        MovingAverageButtom = tk.Button(self, text="Medias Móviles",command=lambda: controller.show_frame("MovingAverageB"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 200, window=MovingAverageButtom, height=50, width=200)
        
        #Boton para llamar a la Regresión Lineal Multiple
        RLM = tk.Button(self, text="Regresion Lineal Multiple",command=lambda: controller.show_frame("RLM_BTC"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 300, window=RLM, height=50, width=200)
        #SVM
        SVMBTC = tk.Button(self, text="Support Vector Machine",command=lambda: controller.show_frame("SVM_BTC"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 400, window=SVMBTC, height=50, width=200)
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)

class Cardano(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='CARDANO')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 40, window=Titulo)
        
        def pplot ():
            Historicobutton.destroy()
            output = FigureCanvasTkAgg(plotloo(), master=self)
            output.get_tk_widget().pack()
            output.draw()
            
            if output:
                def clearplot():
                    limpiarbutton.destroy()
                    output.get_tk_widget().destroy()
                    Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )

                limpiarbutton = tk.Button(self, text="Limpiar",command = lambda:clearplot(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(500, 300,window=limpiarbutton, height=50, width=180)
            else:
                print("error")
                
        def plotMedias (): 
            #la funcion plotloo esta en un script aparte
            outputt = FigureCanvasTkAgg(pplotMediasADA(), master=self)
            #llamada a la funcion plotloob del script HISTCRYPTO
            outputt.get_tk_widget().pack()
            outputt.draw()
            #en el caso de tener el historico
            if outputt:
                #funcion para limpiar la grafica
                def clearplotMedias():
                    limpiarMediasbutton.destroy()
                    outputt.get_tk_widget().destroy()
                    #creo otro boton por si queremos crear, destruir continuamente
                    HistoricoMediabutton = tk.Button(self, text="Medias Móviles en el último año",command =lambda: plotMedias() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(750, 200, window=HistoricoMediabutton,height=50, width=250 )
                    
                    #creo otro boton por si queremos crear, destruir continuamente
                    Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )

                limpiarMediasbutton = tk.Button(self, text="Limpiar",command = lambda:clearplotMedias(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(750, 300,window=limpiarMediasbutton, height=50, width=180)
            else:
                print("error")
                
        HistoricoMediabutton = tk.Button(self, text="Medias Móviles en el último año",command =lambda: plotMedias() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(750, 200, window=HistoricoMediabutton,height=50, width=250 )
                
                
        Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )
        
        #Boton para llamar a la MediaMovil
        MovingAverageButtom = tk.Button(self, text="Medias Móviles",command=lambda: controller.show_frame("MovingAverageA"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 200, window=MovingAverageButtom, height=50, width=200)
        
        #Regresión Lineal Multiple
         #Boton para llamar a la Regresión Lineal Multiple
        RLM = tk.Button(self, text="Regresion Lineal Multiple",command=lambda: controller.show_frame("RLM_ADA"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 300, window=RLM, height=50, width=200)
         #SVM
        SVMADA = tk.Button(self, text="Support Vector Machine",command=lambda: controller.show_frame("SVM_ADA"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 400, window=SVMADA, height=50, width=200)
        
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)
        
        
class Ethereum(tk.Frame):

    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        self.controller = controller
        
        canvas = tk.Canvas(self, width = 1000, height = 500,
                           #bg='#7BE393'
                          )
        canvas.pack()
        
        Titulo = tk.Label(self, text='ETHEREUM')
        Titulo.config(font=('Times New Roman', 36))
        canvas.create_window(500, 40, window=Titulo)
        
        def pplot ():
            Historicobutton.destroy()
            output = FigureCanvasTkAgg(plotlooe(), master=self)
            output.get_tk_widget().pack()
            output.draw()
            
            if output:
                def clearplot():
                    limpiarbutton.destroy()
                    output.get_tk_widget().destroy()
                    Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )

                limpiarbutton = tk.Button(self, text="Limpiar",command = lambda:clearplot(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(500, 300,window=limpiarbutton, height=50, width=180)
            else:
                print("error")
                
        def plotMedias (): 
            #la funcion plotloo esta en un script aparte
            outputt = FigureCanvasTkAgg(pplotMediasET(), master=self)
            #llamada a la funcion plotloob del script HISTCRYPTO
            outputt.get_tk_widget().pack()
            outputt.draw()
            #en el caso de tener el historico
            if outputt:
                #funcion para limpiar la grafica
                def clearplotMedias():
                    limpiarMediasbutton.destroy()
                    outputt.get_tk_widget().destroy()
                    #creo otro boton por si queremos crear, destruir continuamente
                    HistoricoMediabutton = tk.Button(self, text="Medias Móviles en el último año",command =lambda: plotMedias() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(750, 200, window=HistoricoMediabutton,height=50, width=250 )
                    
                    #creo otro boton por si queremos crear, destruir continuamente
                    Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
                    canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )

                limpiarMediasbutton = tk.Button(self, text="Limpiar",command = lambda:clearplotMedias(),
                        bg='brown',fg='white',font=('Arial', 11, 'bold'))
                canvas.create_window(750, 300,window=limpiarMediasbutton, height=50, width=180)
            else:
                print("error")
                
        HistoricoMediabutton = tk.Button(self, text="Medias Móviles en el último año",command =lambda: plotMedias() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(750, 200, window=HistoricoMediabutton,height=50, width=250 )
                
        Historicobutton = tk.Button(self, text="Historico",command =lambda: pplot() ,bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 200, window=Historicobutton,height=50, width=180 )
        
        #Boton para llamar a la MediaMovil
        MovingAverageButtom = tk.Button(self, text="Medias Móviles",command=lambda: controller.show_frame("MovingAverageE"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 200, window=MovingAverageButtom, height=50, width=200)
        
        #Regresión Lineal Multiple
         #Boton para llamar a la Regresión Lineal Multiple
        RLMETH = tk.Button(self, text="Regresion Lineal Multiple",command=lambda: controller.show_frame("RLM_ETH"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 300, window=RLMETH, height=50, width=200)
        
        #SVM
        SVMETH = tk.Button(self, text="Support Vector Machine",command=lambda: controller.show_frame("SVM_ETH"),
                             bg='lightskyblue2', font=('Arial', 11, 'bold'))
        canvas.create_window(150, 400, window=SVMETH, height=50, width=200)
        
        button = tk.Button(self, text="Regresar al Menu",command=lambda: controller.show_frame("Main"),bg='brown',
                          fg='white',font=('Arial', 11, 'bold'))
        canvas.create_window(500, 400, window=button,  height=50, width=180)

####################################################################################################    

#Para inicializar el gui de tkinter en el caso de que se inicialize el Main
if __name__ == "__main__":
    app = SampleApp()
    app.mainloop()

```
