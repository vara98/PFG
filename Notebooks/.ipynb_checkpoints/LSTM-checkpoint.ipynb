```python
%reset

import numpy as np
import pandas as pd
import calendar
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import plotly.express as px
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import time
import warnings
from sklearn.preprocessing import MinMaxScaler
from keras.layers.core import Dense, Activation, Dropout
from keras.layers.recurrent import LSTM
from keras.models import Sequential


df = pd.read_excel('Datos_finanzas/ETHEREUM.xlsx')
temp=pd.DataFrame(df['Fecha'])
nuevacolumna = []
for i in range(0, len(df['Fecha'])):
    nuevacolumna=pd.Timestamp(temp['Fecha'][i]).day_name()
    temp['Fecha'][i]=nuevacolumna 
df=df.assign(Dia_semana=temp)
#Nuevas columnas Dia, Mes y Año
df['Dia'] = pd.DatetimeIndex(df['Fecha']).day
df['Mes'] = pd.DatetimeIndex(df['Fecha']).month
df['Año'] = pd.DatetimeIndex(df['Fecha']).year
#Transformación de la fecha
df['fecha']=df['Año'].apply(str) + "-" + df['Mes'].apply(str) + "-" + df['Dia'].apply(str)
df['fecha'] = pd.to_datetime(df['fecha']) 
df.index = df['fecha']
datos=df
dictionaries = []

for columna in ['Ganancia','Dia_semana']:
    datos[columna], uniques = pd.factorize(datos[columna])
    values = list(range(len(uniques)))
    dictionaries.append(dict(zip(values, uniques)))

datos=datos.drop(columns=['fecha','Fecha'])
datos
datos=df
#Ordeno las fechas
datos = datos.sort_index(ascending=True, axis=0)
#Obtengo los datos para media movil (en este caso utilizo el cierre del mercado y la fecha)
datos=datos.iloc[:,[1,2,3,4,5,6]]
warnings.filterwarnings("ignore")
```

```python
datos
```

```python
#normalizacion
sc = MinMaxScaler(feature_range = (0, 1))
```

```python
#En mi caso quiero coger los datos de cierre por lo que los valores escogidos son 3:4
```

```python
entrenamiento = datos.iloc[:round(len(datos)*0.8), 3:4].values
test = datos.iloc[round(len(datos)*0.8):,3:4].values
```

```python
entrenamiento_norm = sc.fit_transform(entrenamiento)
```

```python
bloque=30
X_train = []
y_train = []
n_dentrenamiento=len(entrenamiento_norm)
for i in range(bloque, n_dentrenamiento):
    X_train.append(entrenamiento_norm[i-bloque:i, 0])
    #X= bloque de 30 datos: 0-30,1-30+1,2-30+1
    y_train.append(entrenamiento_norm[i, 0])
    #y siguente dato
#se almacena los bloques en la posicion de x_train y y_train    
X_train, y_train = np.array(X_train), np.array(y_train)
#hago reshape para que el vector sea (numero de entrenamiento-bloque)x30x1
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
#para predecir 1 dia
```

```python
d_entrada=(X_train.shape[1], 1)
#30x1
d_salida = 1
#numero de neuronas
n_neuronas=500
```

```python
modelo = Sequential()

modelo.add(LSTM(units = n_neuronas, return_sequences = True, input_shape = d_entrada))
modelo.add(Dropout(0.2))



modelo.add(LSTM(units = n_neuronas))
modelo.add(Dropout(0.2))


# Adding the output layer
modelo.add(Dense(units = d_salida))

# Compiling the RNN
modelo.compile(optimizer = 'Adam', loss = 'mse')

# Fitting the RNN to the Training set
modelo.fit(X_train, y_train, epochs = 100, batch_size = 32)
prec=modelo.evaluate(X_train, y_train, verbose=0)
modelo.summary()
```
